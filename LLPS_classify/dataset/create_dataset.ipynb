```python
import pandas as pd
from Bio import SeqIO
import tqdm
```

```python
def fasta_2_csv(fasta_file):
    sequence_id = []
    sequence = []
    for record in SeqIO.parse(fasta_file, "fasta"):
        sequence_id.append(record.id)
        sequence.append(str(record.seq))
    return(pd.DataFrame({'id':sequence_id,'seq':sequence}))
```

```python
llps_record_75 = SeqIO.parse('./cleaned_data/llps_data_cleaned_cdhit75.fasta',"fasta")
llps_record_85 = SeqIO.parse('./cleaned_data/llps_data_cleaned_cdhit85.fasta',"fasta")
llps_record_90 = SeqIO.parse('./cleaned_data/llps_data_cleaned_cdhit90.fasta',"fasta")
```

```python
llps_df_75 = fasta_2_csv('./cleaned_data/llps_data_cleaned_cdhit75.fasta')
llps_df_85 = fasta_2_csv('./cleaned_data/llps_data_cleaned_cdhit85.fasta')
llps_df_90 = fasta_2_csv('./cleaned_data/llps_data_cleaned_cdhit90.fasta')
```

```python
def split_dataset(df,val=False,split_size = 0.2):
    from sklearn.model_selection import train_test_split


    # 随机划分DataFrame
    train_df, test_df = train_test_split(df, test_size=split_size, random_state=42)
    if(val):
        train_df, val_df = train_test_split(train_df, test_size=0.2, random_state=42)
        return train_df,val_df,test_df
    else:
        return train_df,test_df
```

```python
train_llps_df,test_llps_df = split_dataset(llps_df_75)
```
