```python
def df_protein_clean(df,seq = 'seq',min_length = 5,max_weight = 75):
    df =df[~df[seq].isnull()]
    pattern = re.compile('[^ACDEFGHIKLMNPQRSTVWY]')
    # 删除包含非法字符的行
    df = df[~df[seq].str.contains(pattern)]
    
    df['weight'] = df[seq].apply(lambda x: ProteinAnalysis(x).molecular_weight()/1000 )
    df['len'] = df[seq].apply(lambda x: len(x))
    df = df[df['len']>min_length]
    df = df[df['weight']<max_weight]
    
    df = df.drop(columns=['weight','len'])
    return(df)
```

```python
DisPort_data = df_protein_clean(DisPort_data,seq='sequence')
DisPort_data.rename(columns={'sequence':'seq','acc':'id'},inplace=True)
```

```python
Structure_data = df_protein_clean(Structure_data)
Structure_data
```

```python
import os
def df2fasta(df,name,id = 'id',seq = 'seq',dir = None):
    # print('./cleaned_data/{}.csv'.format(name))
    print(dir)
    with open(dir,'w') as file:
        for index, row in df.iterrows():
            file.write('>'+row[id]+'_'+str(row['label'])+'\n'+row[seq]+'\n')
    file.close()
# df2fasta(negative_sample,'negative_sample')
# df2fasta(negative_data,'negative_data')
```

```python
import pandas as pd
for file in os.listdir('./dataset1.0/'):
    print(file)
    if(file.endswith('.csv')):
        df = pd.read_csv(os.path.join('./dataset1.0/',file))
        if file =='llpsdb_d.csv':df['label'] = 1
        df2fasta(df,file,dir='./dataset1.0/{}.fasta'.format(file.split('.')[0]))
```
